```python
import pandas as pd
import numpy as np 
import matplotlib as plt
import seaborn as sns

# Load CSV
# Update the path to the correct location of your CSV file
df = pd.read_csv('startup_funding_sample.csv')  # Make sure the file is in the same directory as your notebook, or update the path accordingly

# Check first 5 rows
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['Funding Amount (Cr)'] = pd.to_numeric(df['Funding Amount (Cr)'], errors='coerce')
```

```python
df["Date"] =pd.to_datetime(df['Date'])
```

EDA

```python


import matplotlib.pyplot as plt
import seaborn as sns

industry_funding = df.groupby('Industry')['Funding Amount (Cr)'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=industry_funding.index, y = industry_funding.values)
plt.title("Top funded Industries")
plt.ylabel("Funding Amount (Cr)")
plt.show()
```

Year wise funding trend 

```python
# year wise funding Trend 
df['Year'] =   df['Date'].dt.year
yearly_funding = df.groupby('Year')['Funding Amount (Cr)'].sum()
yearly_funding.plot(kind='line', marker = 'o', title="Year-wise Funding trend")
```

Top Investors 

```python
investor_funding = df.groupby('Investors')['Funding Amount (Cr)'].sum()
sns.barplot(x=investor_funding.index, y = investor_funding.values)
plt.xticks(rotation = 45)
plt.title("Top Investors")
plt.show()
```

Location-wise Distribution

```python
location_funding = df.groupby('Location')['Funding Amount (Cr)'].sum()
location_funding.plot(kind='pie', autopct = '%1.1f%%', title = 'Funding by Locaction')
plt.show()
```
